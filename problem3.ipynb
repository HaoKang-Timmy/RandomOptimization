```python
import mlrose
import numpy as np
def queens_fitness(state):
    n = len(state)
    attacks = 0
    for i in range(n):
        for j in range(i + 1, n):
            if state[i] == state[j] or abs(i - j) == abs(state[i] - state[j]):
                attacks += 1
    return -attacks  # We want to maximize fitness, so we use negative attacks

# Define the problem
problem = mlrose.DiscreteOpt(length=8,fitness_fn=mlrose.CustomFitness(queens_fitness), maximize=True, max_val=8)
import time
# Define the MIMIC Algorithm
start = time.time()
algorithm = mlrose.algorithms.mimic(problem, pop_size=1000, keep_pct=0.2,curve=True)
end = time.time()
# Run the MIMIC Algorithm
best_state, best_fitness,mimic_curve = algorithm

print("MIMIC - Best state:", best_state)
print("MIMIC - Best fitness:", best_fitness)
print(end - start)
```

```python
import mlrose
import numpy as np

# Define the problem
problem = mlrose.DiscreteOpt(length=8,fitness_fn=mlrose.CustomFitness(queens_fitness), maximize=True, max_val=8)
start = time.time()
# Define the Simulated Annealing Algorithm
algorithm = mlrose.algorithms.simulated_annealing(problem, max_attempts=1000, max_iters=1000,curve=True)
end = time.time()
# Run the Simulated Annealing Algorithm
best_state, best_fitness,sa_curve = algorithm

print("Simulated Annealing - Best state:", best_state)
print("Simulated Annealing - Best fitness:", best_fitness)
print(end-start)
```

```python
import mlrose
import numpy as np

# Define the problem
problem = mlrose.DiscreteOpt(length=8, fitness_fn=mlrose.CustomFitness(queens_fitness), maximize=True, max_val=8)

# Define the Random Hill Climbing Algorithm
start = time.time()
algorithm = mlrose.algorithms.hill_climb(problem,  max_iters=1000,curve=True)
end = time.time()
# Run the Random Hill Climbing Algorithm
best_state, best_fitness,hc_curve = algorithm

print("Random Hill Climbing - Best state:", best_state)
print("Random Hill Climbing - Best fitness:", best_fitness)
print(end-start)
```

```python
import mlrose
import numpy as np

# Define the problem
problem = mlrose.DiscreteOpt(length=8, fitness_fn=mlrose.CustomFitness(queens_fitness), maximize=True, max_val=8)

# Define the Genetic Algorithm
start = time.time()
algorithm = mlrose.algorithms.genetic_alg(problem, pop_size=1000, mutation_prob=0.1,curve=True)
end = time.time()
# Run the Genetic Algorithm
best_state, best_fitness,ga_curve = algorithm

print("Genetic Algorithm - Best state:", best_state)
print("Genetic Algorithm - Best fitness:", best_fitness)
print(end - start)
```
